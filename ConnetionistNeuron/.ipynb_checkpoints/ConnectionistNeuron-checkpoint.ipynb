# Connectionist Neuron

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Load the dataset and plot the data as scatter plot

```python
data = pd.read_csv('applesOranges.csv')
print (data.head())
datanp = np.array(data)

apples = np.array(data[data['y'] == 0])
oranges = np.array(data[data['y'] == 1])
plt.figure(figsize = (8, 6))
plt.scatter(apples[:, 0], apples[:, 1], label = 'apples')
plt.scatter(oranges[:, 0], oranges[:, 1], label = 'oranges')
plt.xlabel('x.1'); plt.ylabel('x.2'); plt.title('Apples and Oranges dataset'); plt.legend(); plt.show()
```

### Set the bias θ = 0. Create a set of 19 weight vectors w = (w1,w2) pointing from the origin to the upper semi-circle with radius 1. I.e. if α denotes the angle between the weight vector and the x-axis, for each α = 0, 10, . . . , 180 (equally spaced) such that ||w||2 = 1, w1 ∈ [−1, 1], w2 ∈ [0, 1]. For each of these weight vectors w determine the classification performance ρ (% correct classifications) of the corresponding neuron and plot a curve showing ρ as a function of α.

```python
def signFunction(x, w, b):
    return (np.sign(np.dot(x, w) + b) + 1) / 2

alpha = np.linspace(0, 180, 19)
classificationPerformanceAlpha = []
for i in range(len(alpha)):
    w = np.array([np.cos(np.radians(alpha[i])), np.sin(np.radians(alpha[i]))]).reshape((2,1))
    y = signFunction(datanp[:, 0:2], w, 0)
    accuracy = 0
    for j in range(datanp.shape[0]):
        if y[j] == datanp[j, 2]:
            accuracy = accuracy + 1/datanp.shape[0]
    classificationPerformanceAlpha.append(accuracy)
    
plt.figure(figsize= (14, 4))
plt.plot(alpha, classificationPerformanceAlpha)
plt.xlabel('alpha'); plt.ylabel('accuracy'); plt.title('classification performance'); plt.show()
```

### From these weight vectors, pick the w yielding best performance. Now vary the θ ∈ [−3, 3] and pick the value of θ giving the best performance.

```python
maxAlpha = alpha[np.argmax(np.array(classificationPerformanceAlpha))]
wmaxalpha = np.array([np.cos(np.radians(maxAlpha)), np.sin(np.radians(maxAlpha))]).reshape((2,1))
bias = np.linspace(-3, 3, 7)
classificationPerformanceBias = []
for i in range(bias.shape[0]):
    y = signFunction(datanp[:, 0:2], wmaxalpha, bias[i])
    accuracy = 0
    for j in range(datanp.shape[0]):
        if y[j] == datanp[j, 2]:
            accuracy = accuracy + 1/datanp.shape[0]
    classificationPerformanceBias.append(accuracy)
    
plt.figure(figsize= (14, 4))
plt.plot(bias, classificationPerformanceBias)
plt.xlabel('bias'); plt.ylabel('accuracy'); plt.title('classification performance'); plt.show()
```

### Plot the data points, colored according to the classification corresponding to these parameter values. Plot the weight vector w in the same plot.

```python
maxBias = bias[np.argmax(np.array(classificationPerformanceBias))]
print ('best w = {}, best bias = {}, accuracy = {}'.format(maxAlpha, maxBias, np.max(np.array(classificationPerformanceBias))))
ymax = signFunction(datanp[:, 0:2], wmaxalpha, maxBias)
color = ['magenta', 'green']
for i in range(datanp.shape[0]):
    if ymax[i] == 0:
        plt.scatter(datanp[i, 0], datanp[i, 1], c = color[0])
    else:
        plt.scatter(datanp[i, 0], datanp[i, 1], c = color[1])
plt.arrow(0, 0, wmaxalpha[0][0], wmaxalpha[1][0], head_width=0.2, color = 'b')
plt.xlabel('x.1'); plt.ylabel('x.2'); plt.title('classified data'); plt.show()
```

The weight vector is perpendicular to the decision boudary, the classification depends on the sign of the projection ( if '-' point belongs magenta class - apples, and if '+' data point belongs to oranges). 

### Find the best combination of w and θ by exploring all combinations of α and θ (within a sensible range and with sensible precision) and plotting the performance of all combinations in a heatmap.

```python
alphas = np.arange(0,181, 1)
biases = np.arange(-3, 3.1, 0.05)

acc = np.zeros([alphas.shape[0], biases.shape[0]])
for i in range(alphas.shape[0]):
    w = np.array([np.cos(np.radians(alphas[i])), np.sin(np.radians(alphas[i]))]).reshape((2,1))
    for j in range(biases.shape[0]):
        y = signFunction(datanp[:, 0:2], w, biases[j])
        accuracy = 0
        for k in range(datanp.shape[0]):
            if y[k] == datanp[k, 2]:
                accuracy = accuracy + 1/datanp.shape[0]
        acc[i, j] = accuracy
        
fig = plt.figure(figsize = (8, 5))
heatmap = plt.imshow(acc.T)
fig.colorbar(heatmap)
plt.show()

from numpy import unravel_index
maxIndex = unravel_index(acc.argmax(), acc.shape)
print ('best w = {}, best bias = {}, accuracy = {}'.format(alphas[maxIndex[0]], biases[maxIndex[1]], acc[maxIndex]))
```

With one neuron only linear decision boudary is possible. Since apples and oranges data set can be separated lineary, we got high accuracy (0.92). For data sets that are not lineary separable this method will fail, for example XOR.
