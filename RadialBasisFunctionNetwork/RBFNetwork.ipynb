# Radial basis function network

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from sklearn.cluster import KMeans
%matplotlib inline
```

### Create training data

```python
def generateData(mean1, mean2, var):
    cov = np.array([[var, 0], [0, var]])
    N1 = np.random.multivariate_normal(mean1, cov, 60)
    N2 = np.random.multivariate_normal(mean2, cov, 60)
    data = np.zeros(N1.shape)
    for i in range(N1.shape[0]):
        idx = np.random.uniform()
        if idx<0.5:
            data[i] = N1[i]
        else:
            data[i] = N2[i]
    return data
```

```python
data1 = generateData([0,1], [1,0], 0.1)
data2 = generateData([0,0], [1,1], 0.1)
data = np.concatenate((data1, data2), axis = 0)
labels = np.array([60*[0] + 60*[1]]).T
```

```python
plt.scatter(data[:60, 0], data[:60, 1], label='0')
plt.scatter(data[60:, 0], data[60:, 1], label='1')
plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.title('Training data'); plt.show()
```

### Build a kNN classifier that classifies new data (query points) by voting of the K nearest neighbors from the training set. Thus the electoral committee is selected from the training patterns according their Euclidean distance to the query point. The predicted class is determined by the target values of the majority of those K nearest patterns. Plot the training patterns and the decision boundary (e.g. using a contour plot or a high-resolution image of equidistant query points) in input space for K = 1, 3, 5.

```python
def distance(x1, x2):
    return np.linalg.norm(x1 - x2, axis = 1)

def KNN(data, labels, x, y, k):
    predictedLabels = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(y.shape[0]):
            distances = np.zeros([data.shape[0], 1])
            p = np.array([x[j,i],y[j,i]]).reshape((1,2))
            distances = distance(np.repeat(p, 120, axis = 0), data)
            knn = np.argsort(distances, axis=0)[:k]
            if (np.sum(labels[knn]) > int(k/2)):
                predictedLabels[j,i] = 1
            else:
                predictedLabels[j,i] = 0
    return predictedLabels
```

```python
k = [1,3,5]
for i in range(len(k)):
    x, y = np.meshgrid(np.linspace(-1.2, 2.2, 100), np.linspace(-1.2, 2.2, 100))
    plt.figure()
    p = KNN(data, labels,x,y, k[i])
    plt.pcolor(x, y, p)
    plt.scatter(data[:60, 0], data[:60, 1], label='0')
    plt.scatter(data[60:, 0], data[60:, 1], label='1')
    plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.title('K = {}'.format(k[i])); plt.show()
```

### Implement “Parzen window” classifier. Plot the training patterns and decision boundary for variance = 0.5, 0.1, 0.01. 

```python
def weight(x1, x2, variance):
    return np.exp(-(1/(2*variance)) * np.sum((x1 - x2)**2, axis=1))

def parzen(data, labels, x, y, var):
    predictedLabels = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(y.shape[0]):
            weights = np.zeros([data.shape[0], 1])
            p = np.array([x[j,i],y[j,i]]).reshape((1,2))
            weights = weight(np.repeat(p, data.shape[0], axis = 0), data, var).reshape((data.shape[0],1))
            Z = np.sum(weights, axis=0)
            predictedLabels[j,i] = (1/Z)*np.sum(labels*weights)
            predictedLabels[j,i] =  round(predictedLabels[j,i])
    return predictedLabels
```

```python
var = [0.5, 0.1, 0.01]
for i in range(len(var)):
    x, y = np.meshgrid(np.linspace(-1.2, 2.2, 100), np.linspace(-1.2, 2.2, 100))
    plt.figure()
    p = parzen(data, labels,x,y, var[i])
    plt.pcolor(x, y, p)
    plt.scatter(data[:60, 0], data[:60, 1], label='0')
    plt.scatter(data[60:, 0], data[60:, 1], label='1')
    plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.title('variance = {}'.format(var[i])); plt.show()
```

## “Parzen window” classifier and KNN for 3 classes

```python
cov3 = np.array([[0.05, 0], [0, 0.05]])
data3 = np.random.multivariate_normal([0.5, 0.5], cov3, 60)
dataNew = np.concatenate((data, data3), axis = 0)
labs = np.array([60*[0] + 60*[1] + 60*[2]]).T
labelsNew = to_categorical(labs)

plt.scatter(dataNew[:60, 0], dataNew[:60, 1], label='0')
plt.scatter(dataNew[60:120, 0], dataNew[60:120, 1], label='1')
plt.scatter(dataNew[120:, 0], dataNew[120:, 1], label='2')
plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.show()
```

```python
def parzen_3class(data, labels, x, y, var):
    predictedLabels = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(y.shape[0]):
            weights = np.zeros([data.shape[0], 1])
            p = np.array([x[j,i],y[j,i]]).reshape((1,2))
            weights = weight(np.repeat(p, data.shape[0], axis = 0), data, var).reshape((data.shape[0],1))
            Z = np.sum(weights, axis=0)
            labels_coded = (1/Z)*np.sum(labels*weights, axis = 0)
            predictedLabels[j,i] = np.argmax(labels_coded)
    return predictedLabels

var = [0.5, 0.1, 0.01]
for i in range(len(var)):
    x, y = np.meshgrid(np.linspace(-1.2, 2.2, 100), np.linspace(-1.2, 2.2, 100))
    plt.figure()
    p = parzen_3class(dataNew, labelsNew,x,y, var[i])
    plt.pcolor(x, y, p)
    plt.scatter(dataNew[:60, 0], dataNew[:60, 1], label='0')
    plt.scatter(dataNew[60:120, 0], dataNew[60:120, 1], label='1')
    plt.scatter(dataNew[120:, 0], dataNew[120:, 1], label='2')
    plt.xlabel('x1'); plt.ylabel('x2'); plt.legend()
    plt.title('Parzen window classifier, variance = {}'.format(var[i])); plt.show()
```

```python
def KNN_3class(data, labels, x, y, k):
    predictedLabels = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(y.shape[0]):
            distances = np.zeros([data.shape[0], 1])
            p = np.array([x[j,i],y[j,i]]).reshape((1,2))
            distances = distance(np.repeat(p, data.shape[0], axis = 0), data)
            knn = np.argsort(distances, axis=0)[:k]
            labs = np.sum(labels[knn], axis = 0)/k
            predictedLabels[i,j] = np.argmax(labs)
    return predictedLabels

k = [1,3,5]
for i in range(len(k)):
    x, y = np.meshgrid(np.linspace(-1.2, 2.2, 100), np.linspace(-1.2, 2.2, 100))
    plt.figure()
    p = KNN_3class(dataNew, labelsNew,x,y, k[i])
    plt.pcolor(x, y, p)
    plt.scatter(dataNew[:60, 0], dataNew[:60, 1], label='0')
    plt.scatter(dataNew[60:120, 0], dataNew[60:120, 1], label='1')
    plt.scatter(dataNew[120:, 0], dataNew[120:, 1], label='2')
    plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.title('KNN classifier, K = {}'.format(k[i])); plt.show()
```

## Radial basis function network

```python
def rbf(data, numberOfClusters, var, labels, x, y):
    labels = labels.reshape((1, data.shape[0]))
    kmeans = KMeans(n_clusters=numberOfClusters).fit(data)
    centers = kmeans.cluster_centers_
    d = np.ones([numberOfClusters+1, data.shape[0]])
    for i in range(1,numberOfClusters+1):
        c = centers[i-1, :].reshape((1,2))
        for j in range(data.shape[0]):
            d[i, j] = np.exp(- np.sum((data[j,:] - c)**2) / (2. * var))
    w = np.linalg.pinv(d.dot(d.T)).dot(d).dot(labels.T)
    d1 = np.array([x.flatten(), y.flatten()]).T
    predictData = np.ones([numberOfClusters+1, d1.shape[0]])
    for i in range(1,numberOfClusters+1):
        c = centers[i-1, :].reshape((1,2))
        predictData[i, :] = weight(np.repeat(c, d1.shape[0], axis = 0), d1, var)
    predictLabels = w.T.dot(predictData)
    predictLabels[np.where(predictLabels>=0.5)] = 1
    predictLabels[np.where(predictLabels<0.5)] = 0
    return predictLabels.reshape(x.shape), centers
```

```python
k = [2,3,4] # number of representatives
var = [0.5, 0.8] #kernel widths
x, y = np.meshgrid(np.linspace(-1.2, 2.2, 100), np.linspace(-1.2, 2.2, 100))

for i in range(len(var)):
    for j in range(len(k)):
        labs = np.zeros(x.shape)
        labs, centers = rbf(data, k[j], var[i], labels, x, y)
        plt.pcolor(x, y, labs)
        plt.scatter(data[:60, 0], data[:60, 1], label='0')
        plt.scatter(data[60:, 0], data[60:, 1], label='1')
        plt.scatter(centers[:,0], centers[:,1], s=100)
        plt.title('k = {}, var = {}'.format(k[j], var[i]))
        plt.xlabel('x1'); plt.ylabel('x2'); plt.legend(); plt.show()
```
