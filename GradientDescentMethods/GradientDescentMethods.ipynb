# Comparison of gradient descent methods

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Compare the performance of three learning procedures applied to a simple connectionist neuron with linear output function. (i) Gradient (or steepest) descent with constant learning rate, (ii) steepest descent combined with a line search method to determine the learning rate, and (iii) the conjugate gradient method.

### Gradient descent

```python
X = np.array(([1, 1, 1], [-1., 0.3, 2.]))
y = np.array(([-0.1, 0.5, 0.5]))
H = np.dot(X, X.T)
b = -np.dot(X, y.T)
winit = np.array(([-0.45], [0.2]))
```

```python
def gradient(H, w, b):
    return np.dot(H, w) + b.reshape((2,1))
```

```python
weights = np.array((winit))
w = winit
ni = 0.05
for i in range(50):
    w = w - gradient(H, w, b)*ni
    weights = np.append(weights, w, axis = 1)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 4))
ax[0].plot(weights[0, :], weights[1, :], 'o')
ax[0].set_ylabel('w1'); ax[0].set_xlabel('w0'); ax[0].set_title('Weight vector for all iterations')
ax[1].plot(np.arange(0,51), weights[0, :], label = 'w0')
ax[1].plot(np.arange(0, 51), weights[1, :], label = 'w1')
ax[1].legend(); ax[1].set_ylabel('weight'); ax[1].set_xlabel('iteration'); ax[1].set_title('Convergence speed')
fig.suptitle('Gradient descent')
plt.show()
```

### Stepest descent combined with line search

```python
weights = np.array((winit))
w = winit
g = gradient(H, w, b)
ni = np.dot(g.T, g)/(np.dot(np.dot(g.T, H), g))
for i in range(50):
    g = gradient(H, w, b)
    w = w - g*ni
    weights = np.append(weights, w, axis = 1)
    if (np.dot(np.dot(g.T, H), g)) == 0:
        break
    ni = np.dot(g.T, g)/(np.dot(np.dot(g.T, H), g))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 4))
ax[0].plot(weights[0, :], weights[1, :], 'o')
ax[0].set_ylabel('w1'); ax[0].set_xlabel('w0'); ax[0].set_title('Weight vector for all iterations')
ax[1].plot(np.arange(0,weights.shape[1]), weights[0, :], label = 'w0')
ax[1].plot(np.arange(0, weights.shape[1]), weights[1, :], label = 'w1')
ax[1].legend(); ax[1].set_ylabel('weight'); ax[1].set_xlabel('iteration'); ax[1].set_title('Convergence speed')
fig.suptitle('Line search')
plt.show()
```

### Conjugate gradient

```python
w = winit
g = gradient(H, w, b)
d = -gradient(H, winit, b)
ni = -np.dot(d.T, g)/(np.dot(np.dot(d.T, H), d))
weights = np.array((w))
for i in range(15):
    g = gradient(H, w, b)
    g0 = g
    w = w + d*ni
    g = np.dot(H, w) + b.reshape((2,1))
    beta = -(np.dot(g.T, g)/np.dot(g0.T, g0))
    d = g + beta*d
    weights = np.append(weights, w, axis = 1)
    if (np.dot(np.dot(d.T, H), d)) == 0:
        break
    ni = -np.dot(d.T, g)/(np.dot(np.dot(d.T, H), d))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 4))
ax[0].plot(weights[0, :], weights[1, :], 'o')
ax[0].set_ylabel('w1'); ax[0].set_xlabel('w0'); ax[0].set_title('Weight vector for all iterations')
ax[1].plot(np.arange(0,weights.shape[1]), weights[0, :], label = 'w0')
ax[1].plot(np.arange(0, weights.shape[1]), weights[1, :], label = 'w1')
ax[1].legend(); ax[1].set_ylabel('weight'); ax[1].set_xlabel('iteration'); ax[1].set_title('Convergence speed')
fig.suptitle('Conjugate Gradient')
plt.show()
```

Convergence is slow for gradient descent - 35 iteration. Line search convergenges after 6 iterations. The fastest convergence was achieved for conjugate gradient. Algorithm converged in 2 iteration. The first update is the same for line search and conjugate gradient, but line search needs more iterations near optimum in order to converge.
