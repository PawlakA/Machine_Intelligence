```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
import datetime
import matplotlib.image as mpimg
from sklearn.feature_extraction.image import extract_patches_2d
from sklearn.decomposition import PCA
import os, glob
```

## fastICA vs. Infomax
### Apply fastICA to the two soundfiles data set (once again) of problem sheet 6 and compare runtime and robustness w.r.t. the mixing matrix A with the Infomax-based ICA-algorithm.

```python
def logistic(x):
    return 1 / (1 + np.exp(-x))

def der_ratio(x):
    return 1 - 2 * logistic(x)

def natural(W, x):
    x = np.reshape(x, (2,1))
    ratio = der_ratio(np.dot(W, x))
    delta_W = np.dot(ratio, np.dot(W, x).T)
    delta_W = W + np.dot(delta_W, W) # np.dot(I, W) = W
    for i in range(x.shape[0]):  
        delta_W[i, i] = 0
    return delta_W
```

```python
s1 = pd.read_csv('sounds/sound1.dat', header=None).as_matrix()
s2 = pd.read_csv('sounds/sound2.dat', header=None).as_matrix()
s = np.concatenate([s1,s2],axis=1).T

A = np.linalg.inv(np.random.rand(2,2))
X = A.dot(s)
m = X.mean(axis=1, keepdims=True)
X= X-m

def infomax(x, eps, lam):
    W = np.linalg.inv(np.random.rand(2, 2))
    X_perm= np.random.permutation(X.T).T
#     X_perm = np.tile(X_perm, 5)
    for i in range(X_perm.shape[1]):
        x= X_perm[:, i]
        x= np.reshape(x, (2,1))
        for i in range(W.shape[0]):
            W[i, i] = 1
        eps= eps*lam
        W += eps * natural(W, x)
    return W

W = infomax(X, 0.01, 0.9999)
unmixed_infomax = W.dot(A.dot(s))

def fastICA(X):
    ica = FastICA(n_components=2)
    S_ = ica.fit_transform(X.T)
    return S_.T

unmixed_fast = fastICA(X)

t= np.arange(s.shape[1])/8192
fig1= plt.figure(figsize= (16,2))
plt.suptitle('sources')
fig1.add_subplot(1,2,1)
plt.plot(t, s1); plt.xlabel('time'); plt.ylabel('amplitude')
fig1.add_subplot(1,2,2)
plt.plot(t, s2); plt.xlabel('time'); plt.ylabel('amplitude')
plt.show()

fig2= plt.figure(figsize= (16,2))
plt.suptitle('mixed sources')
fig2.add_subplot(1,2,1)
plt.plot(t, X[0, :]); plt.xlabel('time'); plt.ylabel('amplitude')
fig2.add_subplot(1,2,2)
plt.plot(t, X[1, :]); plt.xlabel('time'); plt.ylabel('amplitude')
plt.show()

fig3= plt.figure(figsize= (16,2))
plt.suptitle('unmixed sources, Infomax')
fig3.add_subplot(1,2,1)
plt.plot(t, unmixed_infomax[0, :]); plt.xlabel('time'); plt.ylabel('amplitude')
fig3.add_subplot(1,2,2)
plt.plot(t, unmixed_infomax[1, :]); plt.xlabel('time'); plt.ylabel('amplitude')
plt.show()

fig4= plt.figure(figsize= (16,2))
plt.suptitle('unmixed sources, fastICA')
fig4.add_subplot(1,2,1)
plt.plot(t, unmixed_fast[0, :]); plt.xlabel('time'); plt.ylabel('amplitude')
fig4.add_subplot(1,2,2)
plt.plot(t, unmixed_fast[1, :]); plt.xlabel('time'); plt.ylabel('amplitude')
plt.show()
```

```python
time_infomax = np.zeros((50))
time_fast = np.zeros((50))

for i in range(50):
    A = np.linalg.inv(np.random.rand(2,2))
    X = A.dot(s)
    m = X.mean(axis=1, keepdims=True)
    X= X-m
    a = datetime.datetime.now()
    W = infomax(X, 0.01, 0.9999)
    b = datetime.datetime.now()
    time = b-a
    time = time.microseconds/1000000
    time_infomax[i] = time
    c = datetime.datetime.now()
    W1 = fastICA(X)
    d = datetime.datetime.now()
    time = d-c
    time = time.microseconds/1000000
    time_fast[i] = time
    
plt.figure()
plt.plot(time_infomax, label= 'Infomax')
plt.plot(time_fast, label= 'fastICA')
plt.xlabel('mixing matrix'); plt.ylabel('runtime')
plt.legend(loc='upper right'); plt.title('runtime [s] for infomax and fastICA')
plt.show()
```

FastICA needs less time than Infomax even if we only iterate over permuted data once. Also fastICA gives better results for different mixing matrices. This means that fastICA is faster and more robust w.r.t. the mixing matrix.

## ICA on Image Patches
### Sample P patches of N  N pixels from all images of each category and rearrange each sample to a column vector. Recommended are P  20000 and N  12

```python
building_files = glob.glob("imgpca/b*.jpg")
nature_files = glob.glob("imgpca/n*.jpg")
text_files = glob.glob("imgpca/t*.jpg")
building = []
nature = []
text = []
for building_file in building_files:
    building.append(mpimg.imread(building_file))
for nature_file in nature_files:
    nature.append(mpimg.imread(nature_file))  
for text_file in text_files:
    text.append(mpimg.imread(text_file))
        
fig, ax = plt.subplots(1, 3, figsize = (20, 5))
ax[0].imshow(nature[6])
ax[0].set_title('Example image from nature category')
ax[1].imshow(building[3])
ax[1].set_title('Example image from building category')
ax[2].imshow(text[3])
ax[2].set_title('Example image from text category')
plt.show()
```

```python
rng = np.random.RandomState(0)

def get_patches(list_images, size, patches_no):
    patches = np.empty((patches_no*len(list_images), size*size))
    for i in range(len(list_images)):
        p = extract_patches_2d(list_images[i], 
                                          (size, size), 
                                          max_patches=patches_no, 
                                          random_state=rng)
        p = np.reshape(p, (patches_no, size*size))
        patches[i*patches_no:(i+1)*patches_no] = p
    return patches

nature_patches = get_patches(nature, 12, 1600)
building_patches = get_patches(building, 12, 1850)
text_patches = get_patches(text, 12, 1500)

print ('Nature: {} patches'.format(nature_patches.shape[0]))
print ('Buildings: {} patches'.format(building_patches.shape[0]))
print ('Text: {} patches'.format(text_patches.shape[0]))
```

```python
def plot_patches(data, title): 
    fig= plt.figure(figsize=(16, 6))
    fig.suptitle(title)
    for i in range(20):
        vec= np.reshape(data[:, i], (12, 12))
        ax= fig.add_subplot(3, 7, i+1)
        ax.imshow(vec, cmap=plt.cm.gray, interpolation='none')
        ax.set_title(i+1)
    fig.tight_layout()

plot_patches(building_patches.T, '20 random building patches')
```

### Calculate the independent features of the image patches (these are the columns of mixing matrix A). Use a fastICA toolbox to compute this matrix. Show the first 20 independent features as (grayscale) image patches by rearranging the vectors into 12  12 matrices. Order the independent features by decreasing negentropy, (such that the first feature has largest (approximated) negentropy etc).

```python
def sorting(data):
    A3 = data**3
    a = np.sum(A3, axis=0)/data.shape[0]
    A4 = data**4
    A2 = data**2
    b = np.sum(A4, axis=0)/data.shape[0] - 3*((np.sum(A2, axis=0)/data.shape[0])**2)
    negentropy = (a**2)/12 + (b**2)/48
    sort = np.argsort(negentropy)[::-1]
    return data[:, sort]
```

```python
ica = FastICA()
ica.fit(nature_patches)
A_ = ica.mixing_
A = sorting(A_)
plot_patches(A, 'independent features nature')
```

```python
ica = FastICA()
ica.fit(building_patches)
A_ = ica.mixing_
A = sorting(A_)
plot_patches(A, 'independent features buildings')
```

```python
ica = FastICA()
ica.fit(text_patches)
A_ = ica.mixing_
A = sorting(A_)
plot_patches(A, 'independent features text')
```

```python
pca = PCA()
pca.fit(nature_patches)
components = pca.components_
plot_patches(components.T, 'principal components nature')
```

```python
pca = PCA()
pca.fit(building_patches)
components = pca.components_
plot_patches(components.T, 'principal components buildings')
```

```python
pca = PCA()
pca.fit(text_patches)
components = pca.components_
plot_patches(components.T, 'principal components text')
```

As expected principal component and independet components are different since this 2 methods exploit different features of the data. PCA is based on second order statistics while ICA employs higher moments. Also there are differences in independent components between nature, building and text data sets.
